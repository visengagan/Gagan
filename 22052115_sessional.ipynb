**Artificial Crime Report (Agentic RAG)**

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
class CrimeReportAnalyzer:
    def __init__(self):
        self.report_data = {
            "case_name": "Greenwood Jewelers Armed Robbery",
            "date": datetime(2025, 3, 25),
            "time": "15:15",
            "location": "Downtown Metro City",
            "suspects": 3,
            "weapons": ["9mm pistol", "9mm pistol", "sawed-off shotgun"],
            "stolen_value": 2500000,
            "recovered_value": 320000,
            "vehicle": {
                "description": "black SUV",
                "license_plate": "XJ4-7H9",
                "found_location": "near 7th Street"
            },
            "evidence": ["DNA", "ballistic"],
            "lead_investigator": "Detective Inspector Laura Hayes",
            "witness_observations": ["Eastern European accent"]
        }

    def display_summary(self):
        """Display a summary of the crime report"""
        print("=== CRIME REPORT SUMMARY ===")
        print(f"Case: {self.report_data['case_name']}")
        print(f"Date: {self.report_data['date'].strftime('%B %d, %Y')}")
        print(f"Time: {self.report_data['time']}")
        print(f"Location: {self.report_data['location']}")
        print(f"\nSuspects: {self.report_data['suspects']} masked individuals")
        print("Weapons Used:")
        for weapon in self.report_data['weapons']:
            print(f"- {weapon}")
        print(f"\nStolen Value: ${self.report_data['stolen_value']:,}")
        print(f"Recovered Value: ${self.report_data['recovered_value']:,}")
        print(f"Vehicle: {self.report_data['vehicle']['description']} (License: {self.report_data['vehicle']['license_plate']})")
        print(f"Found abandoned: {self.report_data['vehicle']['found_location']}")
        print("\nEvidence Being Analyzed:")
        for item in self.report_data['evidence']:
            print(f"- {item}")
        print(f"\nLead Investigator: {self.report_data['lead_investigator']}")
        print("\nWitness Observations:")
        for obs in self.report_data['witness_observations']:
            print(f"- {obs}")

    def calculate_recovery_rate(self):
        """Calculate percentage of stolen value recovered"""
        return (self.report_data['recovered_value'] / self.report_data['stolen_value']) * 100

    def plot_value_comparison(self):
        """Create a bar chart comparing stolen vs recovered values"""
        labels = ['Stolen', 'Recovered']
        values = [self.report_data['stolen_value'], self.report_data['recovered_value']]

        plt.figure(figsize=(8, 6))
        bars = plt.bar(labels, values, color=['red', 'green'])
        plt.title('Stolen vs Recovered Value Comparison')
        plt.ylabel('Amount ($)')

        # Add value labels on top of bars
        for bar in bars:
            height = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2., height,
                    f'${height:,}',
                    ha='center', va='bottom')

        plt.show()

    def analyze_weapons(self):
        """Analyze weapon types used in the crime"""
        weapon_counts = pd.Series(self.report_data['weapons']).value_counts()

        plt.figure(figsize=(8, 6))
        weapon_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)
        plt.title('Weapon Distribution Among Suspects')
        plt.ylabel('')
        plt.show()

    def generate_investigation_timeline(self):
        """Generate a timeline of investigation events"""
        timeline = {
            "March 25, 2025 15:15": "Robbery occurs at Greenwood Jewelers",
            "March 25, 2025 15:30": "Police arrive at scene",
            "March 25, 2025 16:45": "Abandoned SUV located near 7th Street",
            "March 25, 2025 18:20": "$320,000 recovered from vehicle",
            "March 26, 2025": "Forensic analysis begins on DNA and ballistic evidence",
            "March 26, 2025": "Case assigned to Detective Inspector Laura Hayes"
        }

        print("\n=== INVESTIGATION TIMELINE ===")
        for time, event in timeline.items():
            print(f"{time}: {event}")

# Main execution
if __name__ == "__main__":
    analyzer = CrimeReportAnalyzer()

    print("Artificial Crime Report Analysis - Greenwood Jewelers Case\n")
    analyzer.display_summary()

    recovery_rate = analyzer.calculate_recovery_rate()
    print(f"\nRecovery Rate: {recovery_rate:.2f}% of stolen value")

    print("\nVisual Analysis:")
    analyzer.plot_value_comparison()
    analyzer.analyze_weapons()

    analyzer.generate_investigation_timeline()

    print("\nInvestigation continues...")
```

** Genetic Algorithm for Ballistic Missile Trajectory
Optimization in a War Zone**

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.spatial.distance import euclidean

class MissileGA:
    def __init__(self):
        # Battlefield parameters
        self.start_point = np.array([0, 0, 0])
        self.target = np.array([100, 100, 50])
        self.target_tolerance = 5.0  # ε in the formulation

        # Enemy defense zones (spherical)
        self.enemy_zones = [
            {'center': np.array([30, 40, 20]), 'radius': 15},
            {'center': np.array([60, 70, 30]), 'radius': 20}
        ]

        # Physical constraints
        self.max_acceleration = 5.0  # G-force limit (m/s²)
        self.min_altitude = 10
        self.max_altitude = 60
        self.max_turn_angle = np.pi/4  # 45 degrees

        # Fuel cost parameters
        self.base_fuel_cost = 1.0
        self.altitude_cost_factor = 0.05
        self.turn_penalty = 2.0

        # GA parameters (adjustable)
        self.pop_size = 200
        self.generations = 500
        self.crossover_rate = 0.85
        self.mutation_rate = 0.1
        self.elitism_rate = 0.1
        self.tournament_size = 5

        # Penalty weights
        self.zone_penalty = 1000
        self.accel_penalty = 500
        self.altitude_penalty = 300
        self.target_penalty = 2000

    def initialize_population(self, num_waypoints=3):
        """Create initial population of trajectories"""
        population = []
        for _ in range(self.pop_size):
            # Create waypoints with bias toward target
            waypoints = []
            for i in range(1, num_waypoints+1):
                # Linear interpolation with random perturbation
                t = i/(num_waypoints+1)
                base_point = self.start_point + t*(self.target-self.start_point)
                perturbation = np.random.normal(0, 10, 3)
                waypoints.append(base_point + perturbation)
            population.append(np.array(waypoints))
        return population

    def calculate_fitness(self, trajectory):
        """Evaluate fitness of a trajectory"""
        full_path = np.vstack([self.start_point, trajectory, self.target])

        # Calculate base fuel cost
        fuel_cost = 0
        for i in range(len(full_path)-1):
            segment_length = euclidean(full_path[i], full_path[i+1])
            altitude_cost = self.altitude_cost_factor * full_path[i,2]
            fuel_cost += (self.base_fuel_cost + altitude_cost) * segment_length

            # Turning penalty
            if i > 0:
                prev_vec = full_path[i] - full_path[i-1]
                curr_vec = full_path[i+1] - full_path[i]
                angle = np.arccos(np.dot(prev_vec, curr_vec) /
                                (np.linalg.norm(prev_vec) * np.linalg.norm(curr_vec)))
                if angle > self.max_turn_angle:
                    fuel_cost += self.turn_penalty * (angle - self.max_turn_angle)

        # Calculate penalties for constraints
        penalties = 0

        # Enemy zone violation
        for point in full_path:
            for zone in self.enemy_zones:
                if euclidean(point, zone['center']) < zone['radius']:
                    penalties += self.zone_penalty

        # Acceleration constraint
        for i in range(1, len(full_path)-1):
            v1 = full_path[i] - full_path[i-1]
            v2 = full_path[i+1] - full_path[i]
            accel = np.linalg.norm(v2 - v1)  # Simplified acceleration
            if accel > self.max_acceleration:
                penalties += self.accel_penalty * (accel - self.max_acceleration)

        # Altitude constraints
        for point in full_path:
            if point[2] < self.min_altitude:
                penalties += self.altitude_penalty * (self.min_altitude - point[2])
            if point[2] > self.max_altitude:
                penalties += self.altitude_penalty * (point[2] - self.max_altitude)

        # Target hit constraint
        final_error = euclidean(full_path[-1], self.target)
        if final_error > self.target_tolerance:
            penalties += self.target_penalty * final_error

        return fuel_cost + penalties

    def tournament_selection(self, population, fitness):
        """Select parents using tournament selection"""
        selected = []
        for _ in range(2):  # Select 2 parents
            contenders = np.random.choice(len(population), self.tournament_size, replace=False)
            contender_fitness = [fitness[i] for i in contenders]
            winner = contenders[np.argmin(contender_fitness)]
            selected.append(population[winner])
        return selected

    def crossover(self, parent1, parent2):
        """Uniform crossover between two trajectories"""
        if np.random.random() > self.crossover_rate:
            return parent1.copy(), parent2.copy()

        child1, child2 = parent1.copy(), parent2.copy()
        for i in range(len(parent1)):
            if np.random.random() < 0.5:
                child1[i], child2[i] = parent2[i].copy(), parent1[i].copy()
        return child1, child2

    def mutate(self, individual):
        """Apply mutation to an individual"""
        for i in range(len(individual)):
            if np.random.random() < self.mutation_rate:
                # Mutate this waypoint
                mutation = np.random.normal(0, 5, 3)
                individual[i] += mutation
                # Ensure altitude stays within bounds
                individual[i,2] = np.clip(individual[i,2], self.min_altitude, self.max_altitude)
        return individual

    def run_ga(self):
        """Execute the genetic algorithm"""
        num_waypoints = 3
        population = self.initialize_population(num_waypoints)
        fitness = [self.calculate_fitness(ind) for ind in population]

        best_fitness = []
        avg_fitness = []

        for gen in range(self.generations):
            # Elitism: preserve best individuals
            elite_size = int(self.elitism_rate * self.pop_size)
            elite_indices = np.argsort(fitness)[:elite_size]
            new_population = [population[i].copy() for i in elite_indices]

            # Generate offspring
            while len(new_population) < self.pop_size:
                # Selection
                parents = self.tournament_selection(population, fitness)

                # Crossover
                offspring1, offspring2 = self.crossover(parents[0], parents[1])

                # Mutation
                offspring1 = self.mutate(offspring1)
                offspring2 = self.mutate(offspring2)

                new_population.extend([offspring1, offspring2])

            population = new_population[:self.pop_size]  # Ensure population size stays constant
            fitness = [self.calculate_fitness(ind) for ind in population]

            # Track statistics
            best_fitness.append(np.min(fitness))
            avg_fitness.append(np.mean(fitness))

            if gen % 50 == 0:
                print(f"Generation {gen}: Best fitness = {best_fitness[-1]:.2f}")

        # Return best solution
        best_idx = np.argmin(fitness)
        return population[best_idx], best_fitness, avg_fitness

    def visualize_trajectory(self, trajectory):
        """Plot the trajectory in 3D space"""
        full_path = np.vstack([self.start_point, trajectory, self.target])

        fig = plt.figure(figsize=(12, 9))
        ax = fig.add_subplot(111, projection='3d')

        # Plot trajectory
        ax.plot(full_path[:,0], full_path[:,1], full_path[:,2],
               'b-o', linewidth=2, markersize=6, label='Optimal Path')

        # Plot start and target
        ax.plot([self.start_point[0]], [self.start_point[1]], [self.start_point[2]],
               'go', markersize=10, label='Launch')
        ax.plot([self.target[0]], [self.target[1]], [self.target[2]],
               'ro', markersize=10, label='Target')

        # Plot enemy zones
        for i, zone in enumerate(self.enemy_zones):
            u, v = np.mgrid[0:2*np.pi:20j, 0:np.pi:10j]
            x = zone['center'][0] + zone['radius'] * np.cos(u) * np.sin(v)
            y = zone['center'][1] + zone['radius'] * np.sin(u) * np.sin(v)
            z = zone['center'][2] + zone['radius'] * np.cos(v)
            # Only label the first zone to avoid duplicate labels
            ax.plot_surface(x, y, z, color='r', alpha=0.2, label='Enemy Zone' if i == 0 else "")

        ax.set_xlabel('X Coordinate')
        ax.set_ylabel('Y Coordinate')
        ax.set_zlabel('Altitude')
        ax.set_title('Optimized Missile Trajectory')
        ax.legend()
        plt.show()

    def plot_fitness(self, best_fitness, avg_fitness):
        """Plot fitness progression over generations"""
        plt.figure(figsize=(10, 6))
        plt.plot(best_fitness, label='Best Fitness')
        plt.plot(avg_fitness, label='Average Fitness')
        plt.xlabel('Generation')
        plt.ylabel('Fitness (Fuel Cost + Penalties)')
        plt.title('Fitness Progression Over Generations')
        plt.legend()
        plt.grid()
        plt.show()

# Main execution
if __name__ == "__main__":
    ga = MissileGA()

    print("Running Genetic Algorithm for Missile Trajectory Optimization...")
    best_trajectory, best_fitness, avg_fitness = ga.run_ga()

    print("\nOptimization complete!")
    print(f"Final best fitness: {best_fitness[-1]:.2f}")

    # Visualize results
    ga.plot_fitness(best_fitness, avg_fitness)
    ga.visualize_trajectory(best_trajectory)
```
